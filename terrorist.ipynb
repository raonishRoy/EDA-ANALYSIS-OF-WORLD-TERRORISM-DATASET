TO SHOW THE NUMBER OF TERRORIST ACTIVITIES PER YEAR

```python
plt.subplots(figsize=(15,7))
sns.countplot('Year',data=df,palette='tab10')
plt.xticks(rotation=90)
plt.title('Number of terrorist activities per year')
plt.show()
```

TO VISUALIZE THE NUMBER OF CASUALITIES EACH YEAR

```python
plt.subplots(figsize=(15,6))
year_cas=df.groupby('Year').Casualities.sum().to_frame().reset_index()
year_cas.columns=['Year','Casualities']
sns.barplot(x=year_cas.Year, y=year_cas.Casualities,palette='Blues_d')
plt.xticks(rotation=90)
plt.title('Number of Casualities Each Year')
plt.show()
```

TO VISUALIZE THE NUMBER ATTACKS IN EACH COUNTRY

```python
plt.subplots(figsize=(17,6))
country_attacks = df.Country.value_counts()[:16].reset_index()
country_attacks.columns=['Country','Total Attacks']
sns.barplot(x=country_attacks.Country,y=country_attacks['Total Attacks'],palette='OrRd_r')
plt.xticks(rotation=90)
plt.title('Number of Total Attacks in Each Country')
plt.show()
```

TO VISUALIZE WHICH MATERIALS ARE MOST USED BY TERRORIST

```python
fig=plt.figure(figsize=(15,6))
sns.countplot('Attacktype',data=df,order=df['Attacktype'].value_counts().index,palette='deep')
plt.xticks(rotation=90)
plt.title('Most Used Attack Type By Terrorist')
plt.show()
```

TO VISUALIZE WHICH GROUP OF TERRORIST ARE HIGHLY ACTIVE IN ATTACKS

```python
fig=plt.figure(figsize=(15,6))
x=df['Group'].value_counts()
x=x[:10,]

sns.barplot(x.index,x.values)

plt.title('Group With Most Terrorist Attacks')
plt.xlabel('Group')
plt.ylabel('Counts')
plt.xticks(rotation=90)
plt.show()
```

TO VISUALIZE WHAT WERE THE MOST TARGET TYPE BY TERRORIST

```python
fig=plt.figure(figsize=(15,6))

sns.barplot(df['Targettype'].value_counts()[:10].index,df['Targettype'].value_counts()[:10].values)
plt.xticks(rotation=90)
plt.xlabel('Target Type')
plt.ylabel('Counts')
plt.title('Most Targeted Type BY Terrorist')
plt.show()
```
